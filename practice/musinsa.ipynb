```python
import smtplib
from email.mime.text import MIMEText
```

```python
url ='https://www.musinsa.com/search/musinsa/goods?q=살로몬&list_kind=small&sortCode=pop&sub_sort=&page=1&display_cnt=0'#무신사링크
headers = {'User-Agent': 'Mozilla/5.0'}#컴퓨터가 접속하듯이 접속
web = requests.get(url,headers=headers).content
source = BeautifulSoup(web,'html.parser')
```

```python
print(source)#무신사 사이트(검색어 살로몬)에 대한 html출력
```

```python
items_name = source.find_all('p',{'class':'list_info'}) #살로몬 제품명 찾기(90개)
items_price = source.find_all('p',{'class':'price'}) #가격 찾기(90개)
names = [] #제품명 저장 리스트
pre_prices = [] # 가격 저장 리스트(할인 전 가격& 할인 후 가격)
items_dict = {} #제품과 가격 매칭 딕셔너리

for item_name in items_name: # 제품명 전처리
    item_name = item_name.get_text().strip()
    if '/' in item_name:
        names.append(item_name.split('/')[0])
    else: names.append(item_name)
for item_price in items_price:#가격1차 전처리 후 pre 리스트에 저장#ex 167,000원 110,000원
    pre_prices.append(item_price.get_text().strip())
    
```

```python
prices = []
for price in pre_prices : #가격 후처리 #ex 110000
    if len(price.split()) - 1:
        price= price.split()[1]
        price= price[:-1]
        price= price.replace(',','')
        prices.append(price)
    else: 
        price = price[:-1].replace(',','')
        prices.append(price)
print(prices)
print('\n',len(prices))
```
