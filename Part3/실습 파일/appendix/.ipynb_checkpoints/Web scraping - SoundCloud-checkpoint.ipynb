```python
# source.find('HTML íƒœê·¸ëª…', {'class':'í•´ë‹¹ íƒœê·¸ì˜ í´ë˜ìŠ¤ ê°’'}) 
# -> íŠ¹ì • í´ë˜ìŠ¤ ê°’ì„ ê°€ì§„ HTML íƒœê·¸ë¥¼ 1ê°œ ì–»ì–´ì˜µë‹ˆë‹¤ (ì—¬ëŸ¬ê°œì˜ íƒœê·¸ê°€ ë™ì¼í•œ í´ë˜ìŠ¤ê°’ì„ ê°€ì§€ê³  ìˆì„ ì‹œ ê°€ì¥ ì²«ë²ˆì§¸ íƒœê·¸)

# source.find_all('HTML íƒœê·¸ëª…', {'class':'í•´ë‹¹ íƒœê·¸ì˜ í´ë˜ìŠ¤ ê°’'})
# -> íŠ¹ì • í´ë˜ìŠ¤ ê°’ì„ ê°€ì§„ HTML íƒœê·¸ë¥¼ ëª¨ë‘ ì–»ì–´ì˜µë‹ˆë‹¤ (listê°€ ë©ë‹ˆë‹¤.)

name_display = source.find('h3', {'class':'profileHeaderInfo__userName'})

if name_display is not None:
    name_display = name_display.get_text() # HTML íƒœê·¸ì—ì„œ í…ìŠ¤íŠ¸ ê°’ë§Œ êº¼ëƒ…ë‹ˆë‹¤.
    if 'Pro' in name_display: 
        name_display = name_display.replace('Pro', '') # 'gravez Pro' -> 'gravez '
    name_display = name_display.strip() # ì•/ë’¤ì˜ ê³µë°±ì„ ì œê±°í•©ë‹ˆë‹¤.

print(name_display)
```

# 2. Track

```python
# ìˆ˜ì§‘í•  ë°ì´í„°

# diff_year # í˜„ì¬ ë‚ ì§œì™€ì˜ ì°¨ì´ ì¤‘ 'në…„' ë‹¨ìœ„ì˜ ê°’ (ex. 0ë…„ 6ê°œì›” ì „ ì—…ë¡œë“œ -> 0)
# diff_month # í˜„ì¬ ë‚ ì§œì™€ì˜ ì°¨ì´ ì¤‘ 'në…„'ì„ ì œì™¸í•œ ë‚˜ë¨¸ì§€ ê°œì›” ìˆ˜ (ex. 3ë…„ 3ê°œì›” ì „ ì—…ë¡œë“œ -> 3) <- ì›¹ í˜ì´ì§€ í™”ë©´ì—ëŠ” ë…¸ì¶œë˜ì§€ ì•Šìœ¼ë‚˜ ê°€ëŠ¥
# genre # ì˜µì…˜ì´ë¼ ì—†ì„ ìˆ˜ë„ ìˆë‹¤ëŠ” ë‚´ìš©ì„ ë°˜ì˜í–ˆìŠµë‹ˆë‹¤.
# num_play 
# num_like
# num_repost

# track_name : track ì´ë¦„ (ìƒˆë¡œ ì¶”ê°€í–ˆìŠµë‹ˆë‹¤.)
# track_url : track URL (ìƒˆë¡œ ì¶”ê°€í–ˆìŠµë‹ˆë‹¤. ì¶”í›„ comment ìˆ˜ì§‘ì„ ìœ„í•´ í™œìš©ë©ë‹ˆë‹¤.)
```

```python
artist = 'gravez' # artist ì´ë¦„ë§Œ ë°”ê¿”ì£¼ë©´ ë‹¤ë¥¸ ì•„í‹°ìŠ¤íŠ¸ì˜ íŠ¸ë™ ì •ë³´ë¥¼ í¬ë¡¤ë§í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.
url = "https://soundcloud.com/" + artist + '/tracks'
```

```python
service = Service(executable_path=ChromeDriverManager().install()) 
driver = webdriver.Chrome(service=service)

driver.get(url) 
time.sleep(3)

# ì•„ë˜ ì½”ë“œê°€ íŠ¹ì • ì›¹í˜ì´ì§€ì—ì„œ Seleniumì„ í™œìš©í•´ ìŠ¤í¬ë¡¤ ë‹¤ìš´í•˜ëŠ” ì½”ë“œì…ë‹ˆë‹¤.
# ê°€ì¥ ì•„ë˜ê¹Œì§€ ìŠ¤í¬ë¡¤ ë‹¤ìš´ í›„ ë§¤ë²ˆ 3ì´ˆì”© ë¡œë”©ì„ ê¸°ë‹¤ë¦° í›„ ìŠ¤í¬ë¡¤ ë‹¤ìš´ì„ ë§ˆì§€ë§‰ê¹Œì§€ ì§„í–‰í•©ë‹ˆë‹¤.
while True:
    lastHeight = driver.execute_script("return document.body.scrollHeight")
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(3)
    newHeight = driver.execute_script("return document.body.scrollHeight")
    if newHeight == lastHeight:
         break

web = driver.page_source
source = BeautifulSoup(web, 'html.parser')
# print(source) # sourceë¥¼ ê·¸ëŒ€ë¡œ ì¶œë ¥í•  ê²½ìš°, ë‹´ê³  ìˆëŠ” í…ìŠ¤íŠ¸ê°€ ë¬´ì²™ ë§ì•„ì„œ ë²„ë²…ì´ê²Œ ë  ìˆ˜ ìˆìŠµë‹ˆë‹¤.
```

```python
driver.close()
driver.quit()
```

```python
tracks = source.find_all('div', {'class':'sound__content'})
print(len(tracks)) # ì´ íŠ¸ë™ì˜ ìˆ˜
```

```python
track_name = []
diff_year = []
diff_month = []
genre = [] 
num_play = []
num_like = []
num_repost = []
track_url = []

if tracks is not None: # íŠ¸ë™ì´ í•˜ë‚˜ë„ ì—†ì„ ê²½ìš°ë¥¼ ëŒ€ë¹„ 
    for track in tracks:

        # 1) Track ì´ë¦„
        track_name.append(track.find('a', {'class':'soundTitle__title'}).get_text().strip())

        # 2) ì—…ë¡œë“œ ë‚ ì§œì™€ í˜„ì¬ ë‚ ì§œì™€ì˜ ì°¨ì´ (ì—° & ì›”)
        # íŠ¹ì • HTML íƒœê·¸ì— ëŒ€í•´ .get_text()ëŠ” ë‚´ë¶€ í…ìŠ¤íŠ¸ë¥¼ êº¼ë‚´ì§€ë§Œ, .attrsëŠ” íƒœê·¸ì˜ Attributeë“¤ì„ êº¼ë‚¼ ìˆ˜ ìˆìŠµë‹ˆë‹¤. 
        uploaded_date = track.find('time', {'class':'relativeTime'}).attrs['datetime'][:10] # '2020-06-21'
        uploaded_date = datetime.datetime.strptime(uploaded_date, '%Y-%m-%d') # íŒŒì´ì¬ì˜ datetime í˜•ì‹ìœ¼ë¡œ ë³€í™˜
        today_date  = datetime.datetime.now()
        
        # íŒŒì´ì¬ì˜ datetime í˜•ì‹ì„ ê°–ê³  ìˆëŠ” 2ê°œì˜ ë‚ ì§œì— ëŒ€í•´ ì•„ë˜ rrule.rruleì„ í™œìš©í•´ ë…„ë„ ê¸°ì¤€ ì°¨ì´ & ê°œì›” ê¸°ì¤€ ì°¨ì´ë¥¼ ê³„ì‚°í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.
        # ì‹¤ì œ rrule.rruleì˜ ê³„ì‚° ê²°ê³¼ ê°’ì´ ì›¹ì‚¬ì´íŠ¸ ë‚´ í‘œê¸°ëœ ì—°/ì›” ìˆ˜ë³´ë‹¤ 1ì”© í¬ë¯€ë¡œ 1ì”© ì°¨ê°í•´ì¤ë‹ˆë‹¤.
        diff_year.append(len(list(rrule.rrule(rrule.YEARLY, dtstart=uploaded_date, until=today_date))) - 1)
        # ì•„ë˜ì™€ ê°™ì´ 12ë¡œ ë‚˜ëˆˆ ë‚˜ë¨¸ì§€ë¥¼ ê³„ì‚°í•˜ì§€ ì•Šìœ¼ë©´ íŠ¸ë™ ì—…ë¡œë“œ ì¼ìë¶€í„° ì˜¤ëŠ˜ ì¼ìê¹Œì§€ì˜ ì´ ê°œì›” ìˆ˜ë¥¼ ê³„ì‚°í•´ë²„ë¦½ë‹ˆë‹¤.
        diff_month.append((len(list(rrule.rrule(rrule.MONTHLY, dtstart=uploaded_date, until=today_date))) - 1) % 12)

        # 3) ì¥ë¥´ 
        genre_tag = track.find('span', {'class':'soundTitle__tagContent'})

        if genre_tag is not None: # ì¥ë¥´ê°€ ì—†ì„ ê²½ìš°ë¥¼ ëŒ€ë¹„
            genre.append(genre_tag.get_text().strip())
        else:
            genre.append('')

        # 4) ì¬ìƒ íšŸìˆ˜
        num_play.append(track.find('li', {'class':'sc-ministats-item'}).attrs['title'].replace(' plays', ''))

        # 5) like ìˆ˜ (Kë¡œ í‘œê¸°ëœ ê°’ë“¤ì€ plain number í˜•íƒœë¡œ ë³€í™˜í•˜ì˜€ìŠµë‹ˆë‹¤.)
        likes = track.find('button', {'class':'sc-button-like'}).get_text()
        if ('K' in likes) and ('.' in likes): # 10.6K -> 10,600
            likes = likes.replace('.', ',')
            likes = likes.replace('K', '00') 
        elif ('K' in likes) and ('.' not in likes): # 13K -> 13,000
            likes = likes.replace('K', ',000')
        num_like.append(likes)

        # 6) repost ìˆ˜
        num_repost.append(track.find('button', {'class':'sc-button-repost'}).get_text())
        
        # 7) track URL 
        track_simple = track.find('a', {'class':'soundTitle__title'}).attrs['href']
        track_url.append('https://soundcloud.com' + track_simple)
```

```python
# result

# ì•„ë˜ ì½”ë“œë¡œ ê° ë¦¬ìŠ¤íŠ¸ë“¤ì„ Pandas DataFrameì˜ ê° ì—´ë¡œ ë„£ì–´ DataFrameì„ ë§Œë“¤ ìˆ˜ ìˆìŠµë‹ˆë‹¤.
artist_track_df = pd.DataFrame({
    'artist_name' : [artist for i in range(len(track_name))],
    'track_name' : track_name,
    'diff_year' : diff_year,
    'diff_month' : diff_month,
    'genre' : genre,
    'num_play' : num_play,
    'num_like' : num_like,
    'num_repost' : num_repost,
    'track_url' : track_url    
})

artist_track_df
```

```python
# DataFrameì€ ì•„ë˜ì™€ ê°™ì€ ì½”ë“œë¡œ ì—‘ì…€ íŒŒì¼ë¡œ ì €ì¥ ê°€ëŠ¥í•©ë‹ˆë‹¤.

artist_track_df.to_excel('artist_track_{}.xlsx'.format(artist), index=False, encoding='utf-8')

# ì—‘ì…€ íŒŒì¼ì„ ì—´ê³ ì í•  ë•Œ í™œìš©í•  ìˆ˜ ìˆëŠ” ì½”ë“œ
# df = pd.read_excel('artist_track_gravez.xlsx', encoding='utf-8')
# df
```

```python
# ìˆ˜ì§‘í•  ë°ì´í„°

# track_name 
# num_comments # í™”ë©´ì— ì í˜€ìˆëŠ” ì½”ë©˜íŠ¸ ìˆ˜(61)ë³´ë‹¤ ì‹¤ì œ í™”ë©´ì— ì¶œë ¥ë˜ëŠ” ì½”ë©˜íŠ¸ ìˆ˜(50)ê°€ ì¢…ì¢… ì ì€ ê²ƒ ê°™ìŠµë‹ˆë‹¤. (ì‚­ì œëœ ì½”ë©˜íŠ¸ í¬í•¨ í˜¹ì€ ìˆ¨ê¹€ì²˜ë¦¬ëœ ì½”ë©˜íŠ¸ ë“±)

# comment_text # ì½”ë©˜íŠ¸ì— ëŒ€í•œ ë¦¬ì½”ë©˜íŠ¸ ì—­ì‹œë„ íŠ¸ë™ì— ëŒ€í•œ ì½”ë©˜íŠ¸ë¼ê³  íŒë‹¨ë˜ë¯€ë¡œ ì¼ë°˜ ì½”ë©˜íŠ¸ë¡œì„œ ë™ì¼í•˜ê²Œ ìˆ˜ì§‘í•˜ì˜€ìŠµë‹ˆë‹¤. (ë³„ë„ë¡œ ë‹¤ë¥¸ í˜•íƒœë¥¼ ê°–ê³  ìˆì§€ ì•ŠìŠµë‹ˆë‹¤.)
# comment_stamp
# comment_diff_year
# comment_diff_month
```

```python
track_url_sample = track_url[0] # https://soundcloud.com/gravez/swang
track_url_sample

# track_url_sample = 'https://soundcloud.com/gravez/where-we-from'
# track_url_sample
```

```python
service = Service(executable_path=ChromeDriverManager().install()) 
driver = webdriver.Chrome(service=service)

driver.get(track_url_sample) 
time.sleep(3)

while True:
    lastHeight = driver.execute_script("return document.body.scrollHeight")
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(3)
    newHeight = driver.execute_script("return document.body.scrollHeight")
    if newHeight == lastHeight:
         break

web = driver.page_source
source = BeautifulSoup(web, 'html.parser')
# print(source) # sourceë¥¼ ê·¸ëŒ€ë¡œ ì¶œë ¥í•  ê²½ìš°, ë‹´ê³  ìˆëŠ” í…ìŠ¤íŠ¸ê°€ ë¬´ì²™ ë§ì•„ì„œ ë²„ë²…ì´ê²Œ ë  ìˆ˜ ìˆìŠµë‹ˆë‹¤.
```

```python
driver.close()
driver.quit()
```

```python
track_name = source.find('span', {'class':'soundTitle__title'}).get_text().strip()
num_comments = source.find('span', {'class':'commentsList__actualTitle'}).get_text().replace(' comments', '')

print(track_name)
print(num_comments)
```

```python
comments = source.find_all('div', {'class':'commentItem'}) # ê° ì½”ë©˜íŠ¸ë“¤ì´ div íƒœê·¸ ë°•ìŠ¤ ì•ˆì— ë‹´ê²¨ìˆìŠµë‹ˆë‹¤.
# print(len(comments))

comment_text = []
comment_stamp = []
comment_diff_year = []
comment_diff_month = []

if len(comments) != 0: # ì½”ë©˜íŠ¸ê°€ í•˜ë‚˜ë„ ì—†ì„ ê²½ìš°ë¥¼ ëŒ€ë¹„ 
    for comment in comments:
        # 1) ì½”ë©˜íŠ¸ ë‚´ìš©
        comment_text.append(comment.find('div', {'class':'commentItem__body'}).get_text().strip())

        # 2) Timestamp (íŠ¸ë™ ë‚´ ì‹œê°„)
        comment_stamp.append(comment.find('a', {'class':'commentItem__timestampLink'}).get_text())

        # 3) ì½”ë©˜íŠ¸ ë“±ë¡ ë‚ ì§œì™€ í˜„ì¬ ë‚ ì§œì™€ì˜ ì°¨ì´ (ì—° & ì›”) : ìœ„ Trackì—ì„œì˜ ì½”ë“œì— ëŒ€í•œ ì„¤ëª…ì„ ì°¸ê³ í•´ì£¼ì„¸ìš”.
        commented_date = comment.find('time', {'class':'relativeTime'}).attrs['datetime'][:10]
        commented_date = datetime.datetime.strptime(commented_date, '%Y-%m-%d')
        today_date  = datetime.datetime.now()

        comment_diff_year.append(len(list(rrule.rrule(rrule.YEARLY, dtstart=commented_date, until=today_date))) - 1)
        comment_diff_month.append((len(list(rrule.rrule(rrule.MONTHLY, dtstart=commented_date, until=today_date))) - 1) % 12)
```

```python
# result

print('Track :', track_name)
print('# of comments :', num_comments)
print('# of comments (real) :', len(comments))

track_comment_df = pd.DataFrame({
    'comment_text' : comment_text,
    'comment_stamp' : comment_stamp,
    'comment_diff_year' : comment_diff_year,
    'comment_diff_month' : comment_diff_month
})

track_comment_df
```

```python
track_comment_df.to_excel('track_comment_{}_{}.xlsx'.format(artist, track_name), index=False, encoding='utf-8')
```
