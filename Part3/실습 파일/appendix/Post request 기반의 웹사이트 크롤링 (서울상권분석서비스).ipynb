## 4. Form data를 포함하여 URL 과 함께 POST 요청 후 돌아온 데이터를 처리

```python
data = {
    'stdrYyCd' : '2021',
    'stdrSlctQu' : 'sameQu',
    'stdrQuCd' : '4',
    'stdrMnCd' : '202112',
    'selectTerm' : 'quarter',
    'svcIndutyCdL' : 'CS000000',
    'svcIndutyCdM' : 'all',
    'stdrSigngu' : '11',
    'selectInduty' : '1',
    'infoCategory' : 'rent'
}
```

```python
response = requests.post('https://golmok.seoul.go.kr/region/selectRentalPrice.json', data=data).content
```

```python
result = json.loads(response)
result
```
