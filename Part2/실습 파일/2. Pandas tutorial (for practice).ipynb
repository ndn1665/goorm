#### 4. DataFrame 다루기 with Advanced functions

```python
# 기존 df 에서 일부 열만 떼네어 새로운 df 만들기

df_new = df[ ['name', 'hair', 'feathers', 'eggs', 'milk', 'type'] ] # 에러를 고쳐보세요!
```

```python
df_new.head()
```

```python
# hair 열에 1을 더한 새로운 열을 new_hair 로 추가하기

df_new['new_hair'] = df_new['hair'].apply(lambda x : x+1) # '적용하다' 를 영어로?
df_new.head()
```

```python
# Type 을 index 로 하는 새로운 df 만들기 (엑셀의 pivot table 과 유사)

pivot_df = pd.pivot_table(df_new, index='type', aggfunc=np.sum) # Shift+Tab!
#type이 같은 (type기준으로 같은) 행들은 eggs라든지 feathers등 과 같은 다른 요소들은 sum 해버린다.
```

```python
pivot_df
```

```python
# 열 삭제하기

? pivot_df['new_hair'] # '삭제하다'의 앞 3글자
pivot_df
```

```python
# 행 삭제하기 

pivot_df = pivot_df.?([3]) # Database 에서 data point 를 drop!
pivot_df
```

```python
# 열 이름 바꾸기

# '이름을 바꿔주다'?
pivot_df.?(columns = {'eggs':'산란', 'feathers':'깃털'}, inplace=True) # inplace 옵션 == 덮어쓰기 여부

pivot_df.head()
```

```python
# 산란 수 기준으로 정렬하기

pivot_df.?(by='산란', inplace=True) # 내용(value)을 기준으로 정렬(sort), inplace=True : 덮어쓰기
pivot_df.head()
```

```python
# 산란 수 기준으로 내림차순 정렬하기

pivot_df.sort_values(by='산란', ? =False, inplace=True) 
pivot_df.head()
```

```python
# 얕은 복사와 깊은 복사

pivot_df_2 = pivot_df # shallow copy
pivot_df_3 = pivot_df.copy() # deep=True, deep copy
```

```python
del pivot_df['산란']
```

```python
pivot_df_2.head(3) # 원본 DataFrame인 pivot_df와 연결되어 있습니다.
```
